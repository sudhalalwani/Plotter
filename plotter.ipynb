```python
def plotter1(plot):
    if(plot == 1):
        graph = df.iplot(kind = 'line')
    elif(plot == 2):
        graph = df.iplot(kind='scatter')
    elif(plot == 3):
        graph = df.iplot(kind='bar')
    elif(plot == 4):
        graph = df.iplot(kind='hist')
    elif(plot == 5):
        graph = df.iplot(kind='box')
    elif(plot == 6):
        graph = df.iplot(kind='surface')
    else:
            print("please select from 1-6")
```
